## Modificación de datos.

```python
import pandas as pd
```

```python
def movecol(df, cols_to_move=None, ref_col="", place="After"):
    if cols_to_move is None:
        cols_to_move = []

    cols = df.columns.tolist()
    if place == "After":
        seg1 = cols[: list(cols).index(ref_col) + 1]
        seg2 = cols_to_move
    if place == "Before":
        # seg1 = cols[:list(cols).index(ref_col)]
        seg2 = cols_to_move + [ref_col]

    seg1 = [i for i in seg1 if i not in seg2]
    seg3 = [i for i in cols if i not in seg1 + seg2]

    return df[seg1 + seg2 + seg3]
```

```python
df = pd.read_csv("data/InventarioInicial.csv", delimiter="	", encoding="utf-8")
df.info()
df
```

```python
df_rappi = df[df["Tipo"].str.contains("rappi")]
df_tienda = df[~(df["Tipo"].str.contains("rappi"))]
df_rappi
```

```python
print(
    "Duplicados de tienda: ",
    df_tienda["Nombre"].duplicated(keep=False).sum(),
    "\nDuplicados de rappi: ",
    df_rappi["Nombre"].duplicated(keep=False).sum(),
)
```

```python
df_rappi.loc[:, "Nombre"] = (
    df_rappi.loc[:, "Nombre"].str.replace(" rappi", "").str.strip()
)
df_tienda.loc[:, "Nombre"] = (
    df_tienda.loc[:, "Nombre"].str.replace(" rappi", "").str.strip()
)
df_rappi
```

```python
df_rappi = df_rappi.set_index("Nombre")
df_tienda = df_tienda.set_index("Nombre")
df_rappi.to_csv("data/Rappi.csv")
df_tienda.to_csv("data/Tienda.csv")
df_inventario = pd.concat([df_tienda, df_rappi], axis=1)
df_inventario
```

```python
df_inventario.columns = [
    "Tipo Tienda",
    "Piezas Tienda",
    "Precio Tienda",
    "Tipo Rappi",
    "Piezas Rappi",
    "Precio Rappi",
]
df_inventario
```

```python
df_piezas = df_inventario[["Piezas Tienda", "Piezas Rappi"]].T.ffill().bfill().T
df_piezas
```

```python
df_inventario["Piezas Tienda"] = df_piezas["Piezas Tienda"]
df_inventario = df_inventario.drop(columns=["Piezas Rappi"]).rename(
    columns={"Piezas Tienda": "Piezas"}
)
df_inventario
```

```python
df_inventario = movecol(
    df_inventario, ["Piezas"], ref_col="Tipo Tienda", place="Before"
)
df_inventario
```

```python
df_inventario = df_inventario.sort_index()
df_inventario
```

```python
df_inventario = df_inventario.astype({"Piezas": int})
df_inventario
```

```python
df_inventario.to_csv("data/Inventario-OlgaCervera.csv")
```
